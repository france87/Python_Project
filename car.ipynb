##### Number of null values per column


```python
car.isnull().sum()
```

#### Replace null value with the mean

```python
car['Cylinders'].fillna(car['Cylinders'].mean(), inplace=True)
```

#### Find distinct value for column Make

```python
car['Make'].unique()
```

#### Number of Make

```python
car['Make'].value_counts()
```

```python
car.head()
```

#### Filtering Asia and Europe makers

```python
car[car['Origin'].isin(['Asia', "Europe"])]
```

#### Eliminate car with the weight above 4000 (~)

```python
# Note the symbol ~ allows to elimante all the dataframe above 4000
car[~(car['Weight'] > 4000)]
```

#### Add three digits to column MPG_City

```python
# Note how to use correctly the lambda function
car['MPG_City'] = car['MPG_City'].apply(lambda x:x + 3)
```
